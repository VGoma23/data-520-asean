```python
# These are the earth engine and google earth engine map libraries needed to import
import ee
import geemap
```

```python
# verifies who you are accessing the earth engine. You might need to sign up for earth engine access if you havent already
ee.Authenticate()
```

```python
# initializes this specific project. I have already shared access to this cloud project with everyone
ee.Initialize(project='iuu-fishing-detections-asean')
```

```python
# Map of ASEAN Member states
m = geemap.Map(center=[15, 120], zoom=4, height=600)
m
```

```python
VIIRS = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMCFG")
VIIRS_image_median_ = VIIRS.median()

vis = {
    "bands": ["avg_rad"],
}

#m.set_center(lat=15, lon=120, zoom=4)
m.add_layer(VIIRS_image_median_, vis, "VIIRS")
m
```

```python
vietnam_coast = m.user_roi # sets the geometry to be equal to the polygon drawn by the user in the map above

VIIRS_vietnam = VIIRS.filterBounds(vietnam_coast).filter(ee.Filter.date('2025-01-01', '2025-3-31'))

VIIRS_vietnam_image = VIIRS_vietnam.sort('system:time_start', False).first().clip(vietnam_coast)

vietnam = geemap.Map(center=[9.4, 107.2], zoom=9.4, height=600)
vietnam.add_layer(VIIRS_vietnam_image, vis, "VIIRS_vietnam")
vietnam
```

```python
VIIRS_vietnam_image
```

```python
min = -1.5
max = 340573 #estimated from documentation
VIIRS_vietnam_image_calibrated = VIIRS_vietnam_image.select('avg_rad').multiply(0.1) # range from 0-1

vis2 = {
    "bands": ["avg_rad"],
    min: 0,
    max: 1,
}

threshold = 0.1
ship_masks = VIIRS_vietnam_image_calibrated.gt(threshold)
ship_masks = ship_masks.selfMask()
vietnam2 = geemap.Map(center=[9.4, 107.2], zoom=9.4, height=600)
vietnam2.add_layer(VIIRS_vietnam_image_calibrated, vis2, "VIIRS_vietnam")
vietnam2.add_layer(ship_masks, {"palette": "red"}, "ship_masks")
vietnam2
#
```
